```python
# Etapa 1 - Listas
maquinas = [
    ["Fresadora", "operando", 65.0, "04/11/2025"],
    ["Torno Convencional", "parada", 25.5, "02/11/2025"],
    ["Milling Machine", "operando", 60.0, "06/11/2025"],
    ["Serra de Fita", "parada", 18.0, "03/11/2025"]
]
```

```python
# Etapa 2 - Dicionário
historico_manutencao = {
    "Fresadora": [
        "Troca de correia - 20/10/2025",
        "Lubrificação geral - 01/11/2025"
    ],
    "Torno Convencional": [
        "Ajuste de folga da bucha - 15/10/2025",
        "Troca de óleo - 30/10/2025"
    ],
    "Milling Machine": [
        "Inspeção elétrica - 22/10/2025",
        "Limpeza de guias - 05/11/2025"
    ],
    "Serra de Fita": [
        "Troca da lâmina - 10/10/2025",
        "Aperto da estrutura - 01/11/2025"
    ]
}


# Função para adicionar manutenção ao histórico
def adicionar_manutencao(nome_maquina, descricao):
    if nome_maquina not in historico_manutencao:
        historico_manutencao[nome_maquina] = []  # cria lista nova
    historico_manutencao[nome_maquina].append(descricao)


# Teste
adicionar_manutencao("Compressor de Ar", "Troca de filtro - 06/11/2025")
adicionar_manutencao("Fresadora", "Ajuste do cabeçote - 07/11/2025")

historico_manutencao
```

```python
#Etapa 3 - String
from datetime import datetime
maquinas = [
    ["Fresadora", "operando", 65.0, "04/11/2025"],
    ["Torno Convencional", "parada", 25.5, "02/11/2025"],
    ["Milling Machine", "operando", 60.0, "06/11/2025"],
    ["Serra de Fita", "parada", 18.0, "03/11/2025"],
    ["Torno CNC", "parada", 20.0, "01/11/2025"]
]

def registrar_medicao(linha):
    partes = linha.split(",")
    nome = partes[0].strip()
    temperatura = float(partes[1].strip())
    status = partes[2].strip()

    for m in maquinas:
        if m[0] == nome:
            m[1] = status
            m[2] = temperatura

            # DESAFIO: atualizar a data se estiver em manutenção
            if status.lower() == "em manutenção":
                data_atual = datetime.now().strftime("%d/%m/%Y")
                m[3] = data_atual
            break


# Teste:
registrar_medicao("Torno CNC, 78.5, operando")
maquinas
```

```python
# Etapa 4 - Arquivos
def salvar_dados_maquinas(lista_maquinas, nome_arquivo="dados_maquinas.txt"):
    with open(nome_arquivo, "w") as arq:
        for m in lista_maquinas:
            linha = f"{m[0]};{m[1]};{m[2]};{m[3]}\n"
            arq.write(linha)
    print("Dados salvos em", nome_arquivo)


# Teste
salvar_dados_maquinas(maquinas)

def carregar_dados_maquinas(nome_arquivo="dados_maquinas.txt"):
    maquinas_lidas = []
    try:
        with open(nome_arquivo, "r") as arq:
            for linha in arq:
                partes = linha.strip().split(";")

                # Segurança: garante que existem 4 partes
                if len(partes) != 4:
                    continue

                nome = partes[0]
                status = partes[1]
                temperatura = float(partes[2])
                ultima = partes[3]

                maquinas_lidas.append([nome, status, temperatura, ultima])

    except FileNotFoundError:
        print("Arquivo não encontrado. Salve primeiro.")

    return maquinas_lidas


# Teste:
carregar_dados_maquinas()
```

```python
# Relatório
def gerar_relatorio(nome_arquivo="relatorio_final.txt"):
    # máquina mais quente
    maquina_quente = max(maquinas, key=lambda x: x[2])

    with open(nome_arquivo, "w") as arq:
        arq.write("RELATÓRIO DE MÁQUINAS\n\n")
        arq.write(f"Máquina mais quente: {maquina_quente[0]} ({maquina_quente[2]} °C)\n\n")

        arq.write("Máquinas em manutenção ou paradas:\n")
        for m in maquinas:
            if m[1].lower() in ["em manutenção", "parada"]:
                arq.write(f"- {m[0]} (última manutenção: {m[3]})\n")

        arq.write("\nQuantidade de manutenções registradas:\n")
        for nome, eventos in historico_manutencao.items():
            arq.write(f"- {nome}: {len(eventos)} registro(s)\n")

    print("Relatório gerado em", nome_arquivo)


# Teste
gerar_relatorio()
```
